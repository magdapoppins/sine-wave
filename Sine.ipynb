# Calculating projectile motion along sine wave
Looking at how to find a sine wave in python, visualize it and select values along the way for exampte to be able to create a projectile movements height values.

```python
import matplotlib.pyplot as plt
import numpy as np
```

Set a sample rate (fs) and frequency (f).

```python
fs = 100
f = 2
```

Find the points on the x-axis for plotting.

```python
x = np.arange(fs)
y = np.sin(2*np.pi*f*x/fs)
```

Plot these data using matplotlib.

```python
plt.plot(x, y)
plt.xlabel("Sample rate")
plt.ylabel("Frequency")
plt.stem(x, y, 'r')
plt.show()
```

```python
for i in range(0, 100, 5):
    print(x[i], y[i])
```

## Get one wave
Since our goal is to calculate a projectile path, we only want the positive slice of one wave.

```python
x_2 = np.arange(fs)
y_2 = np.sin(np.pi*x_2/fs)

plt.plot(x_2, y_2)
plt.xlabel("Sample rate")
plt.ylabel("Frequency")
plt.stem(x_2, y_2, 'r')
plt.show()
```

## Get projectile y values corresponding to this wave
Given a start and end position as well as a max height, calculate points on the path.

```python
original_y = 0
max_y = 250
distance_x = 1000

x_projectile = np.arange(distance_x)
y_projectile = np.sin(np.pi*x_projectile/distance_x)

plt.plot(x_projectile, y_projectile)
plt.xlabel("X")
plt.ylabel("Y (%)")
#plt.stem(x_projectile, y_projectile, 'r')
plt.show()
```
